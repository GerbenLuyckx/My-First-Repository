```python
import pandas as pd
```

```python
fp_admin_project_roles = 'admin_project_roles.csv'
fp_admin_project_users = 'admin_project_users.csv'
fp_admin_roles = 'admin_roles.csv'
fp_admin_users = 'admin_users.csv'
fp_admin_project_user_roles = 'admin_project_user_roles.csv'
df1 = pd.read_csv(fp_admin_project_roles)
df2 = pd.read_csv(fp_admin_project_users)
df3 = pd.read_csv(fp_admin_roles)
df4 = pd.read_csv(fp_admin_users)
df5 = pd.read_csv(fp_admin_project_user_roles)

print("ADMIN PROJECT ROLES")
print(df1.head())
print("ADMIN PROJECT USERS")
print(df2.head())
print("ADMIN ROLES")
print(df3.head())
print("ADMIN USERS")
print(df4.head())
print("ADMIN PROJECT USER ROLES")
print(df5.head())
```

```python
df_role = pd.merge(df5, df3, left_on='role_id',right_on='id',how='left')
print(df_role.head())
```

```python
df_role = df_role.drop(['status'], axis=1)
print(df_role.head())
```

```python
role_agg = df_role.groupby('user_id')['name'].agg(lambda x: ', '.join(x)).reset_index()
print(role_agg.head())
```

```python
df4 = pd.merge(df4, role_agg, left_on='id',right_on='user_id', how='left')
print(df4.head())
```

```python
df4 = df4.rename(columns={'name_y':'Project Roles','name_x':'name'})
print(df4.head())
```

```python
df4 = df4.drop(['user_id'],axis=1)
print(df4)
```

```python
fp_result_admin = 'admin_users+roles.xlsx'
try:
    df4.to_excel(fp_result_admin,index=False)
    print(f'Write excel to {fp_result_admin} succeeded')
except Exception as e:
    print(f'An error occurrerd: {e}')
```
